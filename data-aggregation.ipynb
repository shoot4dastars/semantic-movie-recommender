```python
import pandas as pd
```

```python
df1=pd.read_csv('movies.csv')
df1
```

```python
df1.info()
```

```python
df2=pd.read_csv('actors.csv')
df2.head(10)
```

```python
df2.drop(columns='role',axis=1,inplace=True)
df2.head()
```

```python
df2.rename(columns={'name':'actor'},inplace=True)
df2.head()
```

```python
df2=df2.groupby('id').head(5)
df2.tail()
```

```python
df2=df2.groupby('id')['actor'].apply(lambda x: ', '.join(x.astype(str))).reset_index()
df2.head(10)
```

```python
df2.info()
```

```python
df2.tail(5)
```

```python
merged_df= pd.merge(df1,df2,on='id',how='left')
merged_df.info()
```

```python
df3=pd.read_csv('crew.csv')
df3.head()
```

```python
df3=df3[df3['role'].isin(['Director'])]
df3.info()
```

```python
df3.head()
```

```python
df3=df3.groupby(['id', 'role'])['name'].apply(lambda x: ', '.join(x)).unstack(fill_value='').reset_index()
df3.head()
```

```python
merged_df=pd.merge(merged_df,df3,on='id',how='left')
merged_df.head()
```

```python
merged_df.info()
```

```python
df4=pd.read_csv('genres.csv')
df4
```

```python
df4=df4.groupby('id')['genre'].apply(lambda x: ', '.join(x)).reset_index()
df4.info()
```

```python
merged_df=pd.merge(merged_df,df4,on='id',how='left')
merged_df.info()
```

```python
df5=pd.read_csv('languages.csv')
df5.drop(columns='type',inplace=True,axis=1)
df5.info()
```

```python
df5=df5.groupby('id')['language'].apply(lambda x: ', '.join(x)).reset_index()
df5.head()
```

```python
df5.tail()
```

```python
merged_df=pd.merge(merged_df,df5,on='id',how='left')
merged_df.info()
```

```python
df6=pd.read_csv('posters.csv')
df6.head()
```

```python
merged_df=pd.merge(merged_df,df6,on='id',how='left')
merged_df.info()
```

```python
df7=pd.read_csv('themes.csv')
df7=df7.groupby('id')['theme'].apply(lambda x: '.'.join(x)).reset_index()
df7.tail()
```

```python
df7.info()
```

```python
merged_df=pd.merge(merged_df,df7,on='id',how='left')
merged_df.info()
```

```python
merged_df.rename(columns={
    'Director':'director',
    'link':'poster'
}, inplace=True)
merged_df
```

```python
aggregated_movies=merged_df.head(36667)
aggregated_movies.info()
```

```python
aggregated_movies.to_csv('aggregated_movies.csv',index=False)
```
